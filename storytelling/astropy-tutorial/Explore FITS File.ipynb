**Open a FITS File**

We will make sure we have astropy and ipykernel installed first

```python
import astropy
import ipykernel
from astropy.io import fits

```

Next, we select which image we will open

```python
#image is from https://fits.gsfc.nasa.gov/fits_samples.html
fits_file = fits.open('nasa_fits_sample.fits')

print(fits_file)
```

```python
image_data = fits_file[0].data

print(image_data)
```

You can see that there are multiple arrays when we print, so there are multiple images present. Select one only, lets select the first one.

```python
#file has 4 images, select the first
image_data_first=image_data[0].data
print(image_data_first)
```

Next, import matplotlib to visualize the FITS file

```python
import matplotlib.pyplot as plt
```

Plot the image

```python
plt.figure()
plt.imshow(image_data_first, origin='lower')
plt.colorbar()
plt.show()
```

We can see some stars! They are faint put they are there. Let's see more details with LogNorm.

```python
from matplotlib.colors import LogNorm

plt.figure()
plt.imshow(image_data_first, origin='lower', norm=LogNorm())
plt.colorbar()
plt.show()
```

Hmm this is kind of hard to see, let's make it a greyscale.

```python
from matplotlib.colors import LogNorm

plt.figure()
plt.imshow(image_data_first, origin='lower', norm=LogNorm(), cmap='Greys')
plt.colorbar()
plt.show()

#The darker clusters is where there are stars
```

Let's focus on one spot where there are more stars.

```python
#define y coords first, then x (in terms of pixels)
section1 = image_data_first[50:125]
```

```python
plt.figure()
plt.imshow(section1, origin='lower', norm=LogNorm(), cmap='Greys')
plt.colorbar()
plt.show()
```
